```python
Q1=np.percentile(data1['hum'],25,interpolation='midpoint')
Q3=np.percentile(data1['hum'],75,interpolation='midpoint')
IQR=Q3-Q1
print(Q1)
print(Q3)
print(IQR)
upper=np.where(data1['hum']>=(Q3+1.5*IQR))
lower=np.where(data1['hum']<=(Q1-1.5*IQR))
print(upper)
print(lower)
data1.drop(lower[0],inplace=True)
```

```python
sns.boxplot(data1['hum'])
```

```python
x=data1.drop('cnt',axis=1)
y=data1['cnt']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.feature_selection import SelectKBest,chi2
test=SelectKBest(score_func=chi2,k=3)
fit=test.fit(x,y)
```

```python
#summarize scores
np.set_printoptions(precision=10)
print(fit.scores_)
print(x.iloc[0:5,:])
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
rfe=RFE(model,3,verbose=2)
fit=rfe.fit(x,y)
print("num features:%s"%(fit.n_features_))
print("Selected features: %s"%(fit.support_))
print("Feature Ranking: %s"%(fit.ranking_))
```

```python
x1=data1.iloc[:,[2,12,13]]
x1
```
